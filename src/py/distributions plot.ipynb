```python
import json
import numpy as np
import math
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
rookies = {}

with open('../data/roty-metrics.json', 'r') as f:
    rookies = json.load(f)
```

```python
def create_metric_list(metric):
    target_list=[]
    
    for player in rookies["players"]:
        target_list.append(player["metrics"][metric])

    return target_list
```

```python
def plot_distribution(metric):
    data = create_metric_list(metric)
    data.sort()
    # print(data)
    # bins = np.linspace(math.ceil(min(data)),
    #                   math.floor(max(data)),
    #                    10)

    # plt.xlim([min(data)-5, max(data)+5])

    # plt.hist(data, bins=bins, alpha=0.5)
    plt.title(metric)
    std = np.std(data)
    mean = np.mean(data)
    pdf = stats.norm.pdf(data,mean,std)
    plt.plot(data, pdf)
```

```python
plot_distribution("winshare")
```
